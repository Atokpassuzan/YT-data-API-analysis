```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime 


from youtube_data import youtube_search
from youtube_data import parse_publish_timestamp
```

```python
plt.style.use('fivethirtyeight')
```

```python

test = youtube_search("Americanidol")
test.keys()
```

```python
test['tags']
```

```python
test['channelTitle']
```

```python
df = pd.DataFrame(data=test)
df.head()
```

```python
df1 = df[['title','viewCount','channelTitle','commentCount','likeCount','dislikeCount','tags','favoriteCount','videoId','channelId','categoryId','publishedAt']]
df1.columns = ['Title','viewCount','channelTitle','commentCount','likeCount','dislikeCount','tags','favoriteCount','videoId','channelId','categoryId','publishedAt']
df1.head()
```

```python
import numpy as np
numeric_dtype = ['viewCount','commentCount','likeCount','dislikeCount','favoriteCount']
for i in numeric_dtype:
    df1[i] = df[i].astype(int)

dfr = df1[df1['channelTitle']=='American Idol'].sort_values(ascending=False,by='publishedAt')
dfr10 = dfr.head(14)
dfr10
```

```python
dfr10 = dfr10.sort_values(ascending=False,by='viewCount')
plt.bar(range(dfr10.shape[0]),dfr10['viewCount'])
plt.xlabel(range(dfr10.shape[0]),dfr10['Title'],rotation=90)
plt.ylabel('viewCount in 100 millions')

plt.show()
```

```python
a = pd.DataFrame(dfr10['tags'])
a['tags']
```
