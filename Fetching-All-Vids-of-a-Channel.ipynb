```python
# Function for getting all videos of a channel 

def get_channel_videos(channel_id):
    res = youtube.channels().list(id=channel_id,
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None 
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id,
                                   part='snippet',
                                   maxResults=50,
                                   pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
            
    return videos
                                    
        
```

```python
videos = get_channel_videos('UCWv7vMbMWH4-V0ZXdmDpPBA')
```

```python
with open('videos.csv','w') as vid_csv:
    writer = csv.writer(vid_csv,delimiter=',',lineterminator='\n')
    writer.writerow(['Video_Title','Video_ID'])
    for video in videos:
        row = video['snippet']['title'],video['id']
        writer.writerow(row)
    
    
```
